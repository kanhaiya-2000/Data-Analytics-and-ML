# Matplotlib

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df=pd.read_csv('application_test.csv')
df.columns.tolist()
```

```python
df.dropna(inplace=True)
df
```

```python
df.shape
```

```python
grp=df.groupby('OCCUPATION_TYPE')
x=grp['AMT_INCOME_TOTAL'].agg(np.mean)
y=grp['AMT_CREDIT'].agg(np.mean)
```

```python
plt.figure(figsize=(16,5))
plt.xlabel('OCCUPATION TYPE')
plt.ylabel('AMOUNT PER HEAD')
plt.title('OCCUPATION WISE AMOUNT')
plt.xticks(rotation=90)

plt.plot(x,label="TOTAL INCOME")
plt.plot(y,label="CREDIT")

plt.legend(loc="best",shadow=True,fontsize="large")

plt.grid()
```

By the above graph it can be concluded that 'Cleaning Staff' has the lowest and 'IT Staff' has the highest income of all occupations.It can also be observed that Credit Amount for 'It Staff' is lowest and for 'Managers' it is highest.

```python
gp=df.groupby('ORGANIZATION_TYPE')
a=gp['AMT_INCOME_TOTAL'].agg(np.mean)
b=gp['CNT_FAM_MEMBERS'].agg(np.mean)
```

```python
plt.figure(figsize=(16,5))
plt.xlabel("ORGANIZATION TYPE")
plt.ylabel("AVERAGE INCOME PER HEAD")
plt.title("SECTOR WISE INCOMES")
plt.xticks(rotation=90)

plt.plot(a,label="INCOME")
plt.legend(fontsize="large",loc="best")
plt.grid()
```

Observations made:
Lowest Income is in Trade Sector(specifically type 4).
Highest Income is in Industry Sector(specifically type 6)

```python
plt.figure(figsize=(16,5))
plt.xlabel("ORGANIZATION TYPE")
plt.ylabel("AVERAGE NUMBER OF FAMILY MEMBERS")
plt.title("SECTOR WISE NO. OF FAMILY MEMBERS")
plt.xticks(rotation=90)

plt.plot(b,label="NO. OF FAMILY MEMBERS",color="red")
plt.legend(fontsize="large",loc="best")
plt.grid()
```
